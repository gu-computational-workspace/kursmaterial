```python
# imports
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import pandas_datareader.data as web
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
```

##Step 1: Gather Financial Data from WRDS


```python
data_fundamental=pd.read_csv("tf8okgq0vcifwdp4.csv")
```

```python
data_fundamental
```

##Step 2: Get Yahoo Finance Price Data


```python
tickers=['AAPL','TSLA','GOOGL']
data = yf.download(tickers, start='2010-01-01')

# Reshape the DataFrame so tickers are a column
data = data.stack().reset_index()
data = data.rename(columns={'level_1': 'Ticker'})
```

##Step 3: Merge Fundamental Data with Price Data


```python
# Convert the dates to the correct format
data['Date'] = pd.to_datetime(data['Date'])
data_fundamental['datadate'] = pd.to_datetime(data_fundamental['datadate'])

# merge_asof on Date > datadate and matching on Ticker and tic
merged_data = pd.merge_asof(
    data_fundamental.sort_values('datadate'), 
    data.sort_values('Date'), 
    right_on='Date', 
    left_on='datadate', 
    right_by='Ticker', 
    left_by='tic', 
    direction='forward'  
)


merged_data.head()
```

##Step 4: Calculate Financial Ratios


```python
# RoA = Net Income (niq) / Total Assets (atq)
merged_data['RoA'] = merged_data['niq'] / merged_data['atq']

# Calculate Liabilities/Total Assets
merged_data['Liabilities_to_Total_Assets'] = (merged_data['atq'] - merged_data['ceqq']) / merged_data['atq']

# Calculate Size 
merged_data['Size'] = np.log(merged_data['atq'])

# Display the first few rows of the updated dataframe to see the new ratios
merged_data[['Ticker', 'Date', 'RoA', 'Liabilities_to_Total_Assets', 'Size']].head()
```

##Correlations

```python
# Correlation table
correlation_matrix = merged_data[['Adj Close', 'RoA', 'Liabilities_to_Total_Assets', 'Size']].corr()

correlation_matrix
```

##Visualisation

```python
# Create a scatter plot of RoA vs. Stock Price
plt.figure(figsize=(10, 6))

sns.scatterplot(x='RoA', y='Adj Close', hue='Ticker', data=merged_data, palette='tab10', s=100, alpha=0.7)

plt.xlabel('Return on Assets (RoA)')
plt.ylabel('Stock Price (Adj Close)')
plt.title('Return on Assets (RoA) vs. Stock Price by Ticker')

plt.legend(title='Ticker', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show plot
plt.tight_layout()
plt.show()
```
