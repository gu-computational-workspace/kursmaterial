# Exercise 2.1 Stock Price Analysis and Moving Average Calculation

Objective: The objective of this exercise is for you to practice working with lists, loops, and basic logic. You will calculate a simple moving average (SMA) of stock prices and identify which days the stock price was above the moving average.

# WHY? do such an exercise

**Fundamental Programming Practice:**

- Lists and Data Structures: Working with lists helps students understand how to store and manipulate collections of data.
- Loops and Iteration: Calculating a moving average requires iterating over data, which reinforces loop constructs.
- Conditional Logic: Identifying days when the stock price was above the moving average involves applying conditional statements.
- Understanding Financial Concepts:

**Stock Market Analysis:**

- Moving averages are a key tool in analyzing stock trends and making investment decisions.
- Real-World Application: Applying programming skills to financial data makes the learning process more engaging and relevant.

```python
import csv

# Step 1: Read stock prices from the CSV file
filename = './Data/exercises 2.1 to 2.3.csv'
dates = []
closing_prices = []

with open(filename, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        dates.append(row['Date'])
        closing_prices.append(float(row['Close']))  # Extract the closing prices as floats

# Ask the user for the period for the SMA
period = int(input("Enter the period for the simple moving average (SMA): "))

# Calculate the SMA for each day, starting from the period-th day
sma_values = []
for i in range(period, len(closing_prices)):
    sma = sum(closing_prices[i-period:i]) / period
    sma_values.append((dates[i], sma))  # Store date and corresponding SMA

# Compare each day's stock price with the SMA and store the results
higher_than_sma = []
for i in range(period, len(closing_prices)):
    if closing_prices[i] > sma_values[i-period][1]:  # Compare with the SMA
        higher_than_sma.append(dates[i])

# Output the results with closing prices
print("\nDate\t\tClosing Price\tSMA")
for i in range(len(sma_values)):
    date = sma_values[i][0]
    sma = sma_values[i][1]
    # The closing price corresponding to this date
    price = closing_prices[i + period - 1]  # Adjust index for SMA period
    print(f"{date}\t${price:.2f}\t\t${sma:.2f}")

print("\nDays when the stock price was higher than the SMA:")
for date in higher_than_sma:
    print(date)
```

**Increasing SMA Values:**

### Decision-Making Insights:

**Bullish Signals:**

### Risk Management:

**Set Stop-Loss Orders:**

Monitor for Trend Changes:

- Stay vigilant for any signs of the closing price crossing below the SMA, which could signal a trend reversal.

# Exercise 2.2  Stock Price Analysis with Dictionary Storage and Exponential Moving Average (EMA) Calculation
Objective: calculate the Exponential Moving Average (EMA) instead of the Simple Moving Average (SMA) and store stock prices and EMAs together in a dictionary. This exercise reinforces your understanding of lists, dictionaries, and loop-based calculations in a financial context.


```python
import csv
from datetime import datetime

# Step 1: Read stock prices from the CSV file and store in a dictionary
filename = './Data/exercises 2.1 to 2.3.csv'
stock_data = {}

with open(filename, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        date = row['Date']
        closing_price = float(row['Close'])
        stock_data[date] = {'closing_price': closing_price}

# Step 2: Sort dates in chronological order
dates = sorted(stock_data.keys(), key=lambda date: datetime.strptime(date, '%Y-%m-%d'))

closing_prices = [stock_data[date]['closing_price'] for date in dates]

# Step 3: Ask the user for the period for the EMA
period = int(input("Enter the period for the Exponential Moving Average (EMA): "))

# Check if there are enough data points
if period > len(closing_prices):
    print("Error: The period for EMA is longer than the available data.")
    exit()

# Step 4: Calculate the smoothing factor alpha
alpha = 2 / (period + 1)

# Step 5: Initialize the EMA starting point using SMA
initial_sma = sum(closing_prices[:period]) / period
ema_values = [initial_sma]
stock_data[dates[period - 1]]['ema'] = initial_sma

# Step 6: Loop through stock prices to calculate the EMA
for i in range(period, len(closing_prices)):
    ema = (closing_prices[i] - ema_values[-1]) * alpha + ema_values[-1]
    ema_values.append(ema)
    stock_data[dates[i]]['ema'] = ema  # Store EMA in the dictionary

# Step 7: Compare each day's stock price with the EMA
higher_than_ema = [dates[i] for i in range(period, len(dates)) if closing_prices[i] > ema_values[i - period + 1]]

# Step 8: Output the results
print("\nStock prices and corresponding EMAs:")
for date in dates:
    closing_price = stock_data[date]['closing_price']
    ema = stock_data[date].get('ema', None)
    
    if ema is not None:
        print(f"Date: {date}, Closing Price: {closing_price:.2f}, EMA: {ema:.2f}")
    else:
        print(f"Date: {date}, Closing Price: {closing_price:.2f}, EMA: N/A")
```

# Exercise 2.3 Stock Price Analysis with List Comprehension and Error Handling

Objective: In this exercise, you will enhance your Python skills by using list comprehensions to calculate the Simple Moving Average (SMA). You will also introduce basic error handling to ensure the program can manage incorrect or missing user inputs. This exercise combines previous learning on lists and loops with more advanced Python techniques to write cleaner and more robust code.

```python
import csv
import os

# Step 1: Read data from the CSV file with error handling
filename = './Data/exercises 2.1 to 2.3.csv'
stock_data = []
dates = []

try:
    if not os.path.exists(filename):
        raise FileNotFoundError(f"The file '{filename}' does not exist.")
    
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            try:
                date = row['Date']
                closing_price = float(row['Close'])
                dates.append(date)
                stock_data.append(closing_price)
            except ValueError:
                print(f"Data Error: Invalid data on date {row['Date']}. Skipping.")
                continue

except FileNotFoundError as e:
    print(e)
    exit(1)

# Step 2: User input for period with error handling
while True:
    try:
        period = int(input("Enter the period for the Simple Moving Average (SMA): "))
        if period <= 0:
            raise ValueError("The period must be a positive integer.")
        if period > len(stock_data):
            raise ValueError("The period cannot be greater than the number of data points.")
        break
    except ValueError as e:
        print(f"Invalid input: {e}. Please enter a valid positive integer.")

# Step 3: Calculate SMA using list comprehension
sma = [round(sum(stock_data[i-period:i]) / period, 2) for i in range(period, len(stock_data))]
sma_dates = dates[period:]  # Corresponding dates for SMA values

# Step 4: Identify days where the closing price is higher than the SMA using list comprehension
days_above_sma = [i for i in range(period, len(stock_data)) if stock_data[i] > sma[i - period]]

# Step 5: Output the results
print("\nDate\t\tClosing Price\tSMA")
for i in range(period, len(stock_data)):
    date = dates[i]
    close = stock_data[i]
    sma_value = sma[i - period]
    print(f"{date}\t${close:.2f}\t\t{sma_value:.2f}")

print("\nDays where the stock price was higher than the SMA:")
for i in days_above_sma:
    date = dates[i]
    close = stock_data[i]
    sma_value = sma[i - period]
    print(f"Date: {date}, Closing Price: ${close:.2f}, SMA: ${sma_value:.2f}")
```

# Exercise 2.4 Date Range Filtering and Portfolio Performance

Objective: Write a program that reads a CSV file containing daily stock returns for a portfolio. The file contains both trading and non-trading days. Your task is to filter out the non-trading days (weekends) and then calculate the total portfolio return between two user-specified dates. The program should also identify the highest portfolio return within that date range.


```python
import csv
import os
from datetime import datetime

# Read and Parse the CSV File
filename = './Data/exercise 2.4.csv'  # Replace with your CSV file path
portfolio_data = []

if not os.path.exists(filename):
    print(f"Error: The file '{filename}' does not exist.")
    exit()

with open(filename, 'r') as file:
    reader = csv.DictReader(file)
    # Optionally, print the column names to verify they match your dataset
    # print("Column names:", reader.fieldnames)
    for row in reader:
        try:
            date_str = row['Date']
            date = datetime.strptime(date_str, '%Y-%m-%d')
            daily_return = float(row['Return'])  # Corrected key name
            portfolio_data.append({'Date': date, 'Return': daily_return})
        except ValueError:
            print(f"Data Error: Invalid data on date {row.get('Date', 'Unknown')}. Skipping.")
        except KeyError as e:
            print(f"Key Error: {e}. Available keys are: {list(row.keys())}")
            exit()

# Filter Out Non-Trading Days (Weekends)
trading_days = [entry for entry in portfolio_data if entry['Date'].weekday() < 5]

# Check if trading_days is not empty
if trading_days:
    # Determine the earliest and latest dates in the trading data
    earliest_date = min(entry['Date'] for entry in trading_days)
    latest_date = max(entry['Date'] for entry in trading_days)
    print(f"Data is available from {earliest_date.strftime('%Y-%m-%d')} to {latest_date.strftime('%Y-%m-%d')}.")
else:
    print("No trading data available.")
    exit()


# Get User-Specified Date Range
def get_date_input(prompt):
    while True:
        date_str = input(prompt)
        try:
            date = datetime.strptime(date_str, '%Y-%m-%d')
            return date
        except ValueError:
            print("Invalid date format. Please enter the date in YYYY-MM-DD format.")

start_date = get_date_input("Enter the start date (YYYY-MM-DD): ")
end_date = get_date_input("Enter the end date (YYYY-MM-DD): ")

if start_date > end_date:
    print("Error: Start date must be before or equal to end date.")
    exit()

# Step 4: Filter Data Within the Date Range
filtered_data = [entry for entry in trading_days if start_date <= entry['Date'] <= end_date]

if not filtered_data:
    print("No trading data available within the specified date range.")
    exit()

# Step 5: Calculate Total and Cumulative Portfolio Return

# Total Return (Sum of Daily Returns)
total_return = sum(entry['Return'] for entry in filtered_data) * 100  # Convert to percentage

# Cumulative Return (Compounded Daily Returns)
cumulative_return = 1.0
for entry in filtered_data:
    daily_factor = 1 + entry['Return']
    cumulative_return *= daily_factor

cumulative_return = (cumulative_return - 1) * 100  # Convert to percentage

# Step 6: Identify the Highest Portfolio Return
highest_return_entry = max(filtered_data, key=lambda x: x['Return'])
highest_return_date = highest_return_entry['Date'].strftime('%Y-%m-%d')
highest_return_value = highest_return_entry['Return'] * 100  # Convert to percentage

# Step 7: Output the Results
print(f"\nTotal portfolio return from {start_date.strftime('%Y-%m-%d')} to {end_date.strftime('%Y-%m-%d')}: {total_return:.2f}%")
print(f"Cumulative portfolio return: {cumulative_return:.2f}%")
print(f"Highest daily return within this period was on {highest_return_date} with a return of {highest_return_value:.2f}%")
```

## Summary of the Output:

**Data Availability:** The portfolio data covers the period from January 1, 2024, to March 29, 2024.

**Selected Date Range:** You analyzed the entire available period from January 1, 2024, to March 29, 2024.

**Total Portfolio Return:** The portfolio had a total return of -7.94%, indicating an overall loss of 7.94% over the selected period when simply summing daily returns.

**Cumulative Portfolio Return:** The portfolio experienced a cumulative return of -8.02%, reflecting an overall loss of 8.02% over the period when accounting for the compounding of daily returns.

**Highest Daily Return:** The highest daily return during this period was 1.99% on March 27, 2024

# Exercise 2.5 Stock Price Volatility and Date Ranges

Objective: Write a program that calculates the volatility (standard deviation of daily returns) of a stock within a specific date range, based only on trading days (weekdays). Additionally, identify the most volatile week within this period.

```python
import csv
from datetime import datetime, timedelta
import numpy as np

# Step 1: Read stock prices from the modified_stock_prices.csv file
filename = './Data/exercise 2.5.csv'
stock_data = []

try:
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            date = datetime.strptime(row['Date'], '%Y-%m-%d')
            closing_price = float(row['Close'])
            stock_data.append({'date': date, 'closing_price': closing_price})

except FileNotFoundError:
    print(f"Error: The file '{filename}' was not found.")
    exit(1)
except ValueError:
    print("Error: The data in the file is incorrectly formatted.")
    exit(1)

# Get the start and end date from the data in the file
start_date_in_data = stock_data[0]['date']
end_date_in_data = stock_data[-1]['date']

# Tell the user the available date range
print(f"The data file contains stock prices from {start_date_in_data.strftime('%Y-%m-%d')} to {end_date_in_data.strftime('%Y-%m-%d')}.")

# Step 2: User input for start and end date with validation
while True:
    try:
        start_date_str = input(f"Enter the start date (YYYY-MM-DD) within the data range: ")
        end_date_str = input(f"Enter the end date (YYYY-MM-DD) within the data range: ")
        start_date = datetime.strptime(start_date_str, '%Y-%m-%d')
        end_date = datetime.strptime(end_date_str, '%Y-%m-%d')

        if start_date > end_date:
            raise ValueError("Start date cannot be after end date.")
        if start_date < start_date_in_data or end_date > end_date_in_data:
            raise ValueError(f"Dates must be within the data range: {start_date_in_data.strftime('%Y-%m-%d')} to {end_date_in_data.strftime('%Y-%m-%d')}.")
        break
    except ValueError as e:
        print(f"Invalid date input: {e}. Please enter valid dates within the available range.")

# Step 3: Filter the data to include only trading days (weekdays) within the date range
filtered_data = [entry for entry in stock_data
                 if start_date <= entry['date'] <= end_date and entry['date'].weekday() < 5]  # Weekdays only

if not filtered_data:
    print("No trading data available in the specified date range.")
    exit(1)

# Step 4: Calculate daily returns
daily_returns = []
for i in range(1, len(filtered_data)):
    prev_close = filtered_data[i - 1]['closing_price']
    curr_close = filtered_data[i]['closing_price']
    daily_return = (curr_close - prev_close) / prev_close
    daily_returns.append(daily_return)

# Step 5: Calculate the volatility (standard deviation of daily returns)
volatility = np.std(daily_returns)

# Step 6: Break the data into weekly segments and identify the most volatile week
weekly_volatility = []
current_week = []
for entry in filtered_data:
    if entry['date'].weekday() == 0 and current_week:  # Start a new week on Monday
        weekly_returns = [(current_week[i]['closing_price'] - current_week[i-1]['closing_price']) / current_week[i-1]['closing_price']
                          for i in range(1, len(current_week))]
        weekly_volatility.append((current_week[0]['date'], np.std(weekly_returns)))
        current_week = []

    current_week.append(entry)

# Calculate the last week's volatility if there is any data left
if current_week:
    weekly_returns = [(current_week[i]['closing_price'] - current_week[i-1]['closing_price']) / current_week[i-1]['closing_price']
                      for i in range(1, len(current_week))]
    weekly_volatility.append((current_week[0]['date'], np.std(weekly_returns)))

# Find the week with the highest volatility
most_volatile_week = max(weekly_volatility, key=lambda x: x[1])
most_volatile_week_date = most_volatile_week[0].strftime('%Y-%m-%d')
most_volatile_week_volatility = most_volatile_week[1]

# Step 7: Output the results
print(f"\nVolatility (Standard Deviation of Daily Returns) from {start_date_str} to {end_date_str}: {volatility:.4f}")
print(f"Most volatile week starts on {most_volatile_week_date} with a volatility of {most_volatile_week_volatility:.4f}")
```

# 2.5 Common mistakes:

1. Incorrect CSV File Handling:

Issue: Failing to handle file not found errors or incorrect file formatting.

Solution: Use try-except to catch FileNotFoundError and ValueError to handle incorrect file formats.

2. Not Handling Non-Trading Days Properly:

Issue: Students may forget to filter out non-trading days (weekends).
Consequence: Incorrect volatility and return calculations if weekends are included.

Solution: Use weekday() to exclude weekends (Saturday and Sunday).

3. Incorrect Date Input:

Issue: The user may input dates outside the available range or in an incorrect format.

Solution: Validate user input to ensure the dates are within the range and properly formatted.

4. Off-by-One Errors in Daily Return Calculation:

Issue: Misaligning the calculation of daily returns, such as using the wrong index or range.

Solution: Ensure the daily return is calculated correctly between consecutive days.

5. Failing to Handle Empty Date Ranges:

Issue: The filtered data may be empty if no trading days fall within the specified date range.

Solution: Check if the filtered data is empty and handle the case appropriately.

6. Incorrect Weekly Volatility Calculation:

Issue: Students may miscalculate weekly volatility by not properly segmenting the data into weeks.

Solution: Use a Monday-to-Friday loop to break the data into weekly segments.

7. Misunderstanding Volatility (Standard Deviation):

Issue: Confusing volatility with absolute returns or other metrics, leading to incorrect interpretations.

Solution: Ensure that students correctly understand that volatility is the standard deviation of returns.

8. Hardcoding Date Formats:

Issue: Hardcoding specific date formats may lead to issues if the format changes.

Solution: Use datetime.strptime() consistently to handle date parsing.

9. Not Accounting for Incomplete Weeks:

Issue: Failing to handle the last week if it has fewer than five trading days.

Solution: Ensure that incomplete weeks are still calculated for volatility.

11. Unclear Output Formatting:

Issue: The results may be printed without proper formatting, making them hard to interpret.

Solution: Use :.4f for clear and consistent formatting of volatility values.
