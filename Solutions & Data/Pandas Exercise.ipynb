# Exercise 1: Generating Fundamental Financial Data

```python
# Exercise 1: Generating Fundamental Financial Data

# Step 1: Import necessary libraries
import numpy as np  # For numerical operations
import pandas as pd  # For data manipulation
import matplotlib.pyplot as plt  # For plotting
```

```python
# Step 2: Set a random seed for reproducibility
np.random.seed(42)

# Step 3: Define the columns we need
columns = [
    'Company',
    'Stock Price',
    'Total Earnings',
    'Shares Outstanding',
    'Total Debt',
    'Total Equity',
    'Net Income',
    'Total Assets',
    'Dividend'
]
```

```python
# Step 4: Decide on the number of data points (e.g., days)
n = 100  # Number of data points for each company
```

# Define a random seed to ensure reproducibility of the results.

```python
# Step 5: Generate data for Company A

# Generate Stock Prices between $20 and $50
stock_price_a = np.random.uniform(low=20.0, high=50.0, size=n)

# Generate Total Earnings between $1 million and $5 million
total_earnings_a = np.random.uniform(low=1e6, high=5e6, size=n)

# Generate Shares Outstanding between 10 million and 50 million
shares_outstanding_a = np.random.uniform(low=10e6, high=50e6, size=n)

# Generate Total Debt between $500,000 and $2 million
total_debt_a = np.random.uniform(low=5e5, high=2e6, size=n)

# Generate Total Equity between $1 million and $10 million
total_equity_a = np.random.uniform(low=1e6, high=10e6, size=n)

# Generate Net Income between $500,000 and $3 million
net_income_a = np.random.uniform(low=5e5, high=3e6, size=n)

# Generate Total Assets as Total Debt + Total Equity + random variation
total_assets_a = total_debt_a + total_equity_a + np.random.normal(loc=0, scale=1e5, size=n)

# Generate Dividends
div_a = np.random.uniform(low=-1.0, high=2.0, size=n)
```

```python
# Step 6: Create a DataFrame for Company A
data_a = pd.DataFrame({
    'Company': ['A'] * n,  # Assign 'A' to all rows
    'Stock Price': stock_price_a,
    'Total Earnings': total_earnings_a,
    'Shares Outstanding': shares_outstanding_a,
    'Total Debt': total_debt_a,
    'Total Equity': total_equity_a,
    'Net Income': net_income_a,
    'Total Assets': total_assets_a,
    'Dividends': div_a
})
```

```python
# Step 7: Generate data for Company B with different ranges to ensure uniqueness

# Generate Stock Prices between $30 and $60
stock_price_b = np.random.uniform(low=30.0, high=60.0, size=n)

# Generate Total Earnings between $2 million and $6 million
total_earnings_b = np.random.uniform(low=2e6, high=6e6, size=n)

# Generate Shares Outstanding between 5 million and 25 million
shares_outstanding_b = np.random.uniform(low=5e6, high=25e6, size=n)

# Generate Total Debt between $1 million and $4 million
total_debt_b = np.random.uniform(low=1e6, high=4e6, size=n)

# Generate Total Equity between $2 million and $12 million
total_equity_b = np.random.uniform(low=2e6, high=12e6, size=n)

# Generate Net Income between $1 million and $4 million
net_income_b = np.random.uniform(low=1e6, high=4e6, size=n)

# Generate Total Assets as Total Debt + Total Equity + random variation
total_assets_b = total_debt_b + total_equity_b + np.random.normal(loc=0, scale=1e5, size=n)

# Generate Dividends
div_b = np.random.uniform(low=-1.0, high=2.0, size=n)

# Step 8: Create a DataFrame for Company B
data_b = pd.DataFrame({
    'Company': ['B'] * n,  # Assign 'B' to all rows
    'Stock Price': stock_price_b,
    'Total Earnings': total_earnings_b,
    'Shares Outstanding': shares_outstanding_b,
    'Total Debt': total_debt_b,
    'Total Equity': total_equity_b,
    'Net Income': net_income_b,
    'Total Assets': total_assets_b,
    'Dividends': div_a
})
```

```python

# Step 9: Combine the two company datasets into one DataFrame
df = pd.concat([data_a, data_b], ignore_index=True)

# Step 10: View the combined data
print("Combined DataFrame:")
df
```

# More realistic price trend

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Define the number of days to simulate
n_days = 100

# Starting stock price
initial_price = 40.0  # Starting price of $100

# Daily growth rate (e.g., 0.1% per day)
daily_growth_rate = 0.001  # 0.1% daily increase

# Daily volatility (random fluctuation scale)
daily_volatility = 0.005  # 0.5% standard deviation

# Empty list to store stock prices
stock_prices = []

# Generate stock prices with a trend and random fluctuations
for day in range(n_days):
    if day == 0:
        # On the first day, the price is the initial price
        price = initial_price
    else:
        # Get the price from the previous day
        previous_price = stock_prices[day - 1]
        
        # Generate a random shock from a normal distribution
        random_shock = np.random.normal(loc=0, scale=daily_volatility)
        
        # Calculate the new price with growth and random fluctuation
        price = previous_price * (1 + daily_growth_rate + random_shock)
        
    # Append the calculated price to the list
    stock_prices.append(price)

# Create a DataFrame to store the prices
df_price = pd.DataFrame({
    'Day': range(n_days),
    'Stock Price': stock_prices
})

# Plot the stock price over time
plt.figure(figsize=(10, 6))
plt.plot(df_price['Day'], df_price['Stock Price'], color='green', label='Stock Price')
plt.title('Stock Price Over Time')
plt.xlabel('Day')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```

### Min-Max Scaling in Machine Learning

Min-Max scaling is a common technique used in machine learning to normalize the range of independent variables or features of data. It is an essential preprocessing step, especially when different features have varying units and magnitudes, and can significantly improve the performance of certain algorithms.

#### **What is Min-Max Scaling?**

Min-Max scaling transforms the data by scaling features to lie between a given minimum and maximum value, typically between 0 and 1. The formula for Min-Max scaling is:

<code>X_[scaled] = (X - X_[min])/(X_[max] - X_[min])</code>

Where:
- \( X \) is the original value of a feature.
- \( X_{min} \) and \( X_{max} \) are the minimum and maximum values of the feature across the dataset.
- \( X_{scaled} \) is the transformed value, scaled between 0 and 1.

#### **Why is Min-Max Scaling Important in Machine Learning?**

In machine learning, many algorithms are sensitive to the range of data. If features have vastly different ranges, some algorithms may give more weight to features with larger magnitudes, which can skew the model. Min-Max scaling ensures that all features contribute equally to the model by putting them on the same scale.

##### **Examples of Machine Learning Algorithms Sensitive to Scaling**:

1. **Gradient-based algorithms**: 
   - **Logistic Regression**, **Support Vector Machines (SVM)**, and **Neural Networks** rely on gradient-based optimization methods (e.g., gradient descent).
   - These algorithms can perform poorly when features are on different scales because the gradients become steeper or flatter for different features, leading to slower convergence.

2. **Distance-based algorithms**:
   - **K-Nearest Neighbors (KNN)** and **K-Means Clustering** calculate distances between data points. 
   - Features with larger ranges will dominate distance calculations if not scaled. This can result in inaccurate clustering or nearest-neighbor classifications.

3. **Principal Component Analysis (PCA)**:
   - PCA aims to reduce dimensionality by identifying directions (principal components) that maximize the variance. 
   - If one feature has a larger range, it can dominate the variance and distort the results. Scaling ensures all features contribute equally to the variance calculation.

##### **Non-Sensitive Algorithms**:

Some algorithms, such as **decision trees**, **random forests**, and **XGBoost**, do not require scaling because they are based on rule-based approaches and are insensitive to feature magnitudes. However, it is still a best practice to scale data if you are experimenting with multiple algorithms in a machine learning pipeline.

#### **Advantages of Min-Max Scaling**:

1. **Preserves Relationships**: Unlike other scaling techniques (such as **Z-score normalization**, which shifts and scales based on the mean and standard deviation), Min-Max scaling preserves the relationships between data points. It does not distort the relative distances between values, which can be useful in distance-based algorithms.
   
2. **Bounded Data**: Min-Max scaling produces bounded data between 0 and 1 (or a defined range), which can be beneficial when working with algorithms that expect normalized inputs, such as neural networks (where activation functions like Sigmoid or ReLU work better with small values).

#### **Disadvantages of Min-Max Scaling**:

1. **Sensitive to Outliers**: Since Min-Max scaling is based on the minimum and maximum values of the data, it is sensitive to outliers. An extreme value can significantly distort the scaling. For instance, if you have a dataset where most values lie between 0 and 100, but there's one value at 1000, the scaling will be skewed, compressing the range of most of your data.

2. **Not Centered on Mean**: Unlike standardization (Z-score normalization), Min-Max scaling does not center data around the mean. This can be problematic for some machine learning algorithms that assume data is normally distributed and centered around zero.

#### **Key Takeaways**:
- **Min-Max scaling** normalizes data by scaling features to a 0-1 range, ensuring that no feature dominates others due to differing scales.
- It’s essential for algorithms that are sensitive to feature magnitudes, like gradient-based, distance-based, and variance-based algorithms.
- While it preserves the relative relationships between data points, it is sensitive to outliers.
- When used in a machine learning pipeline, it helps improve algorithm performance and convergence, making the learning process more stable and efficient.
