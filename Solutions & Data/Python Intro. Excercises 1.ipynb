# Exercise 1.1 FizzBuzz

Objective: Write a Python program that prints the numbers from 1 to 100. But for multiples of 3, print "Fizz" instead of the number, and for the multiples of 5, print "Buzz." For numbers which are multiples of both 3 and 5, print "FizzBuzz.


```python
# FizzBuzz Exercise

# Loop through numbers 1 to 100
for num in range(1, 101):
    # Check if divisible by both 3 and 5
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    # Check if divisible by 3
    elif num % 3 == 0:
        print("Fizz")
    # Check if divisible by 5
    elif num % 5 == 0:
        print("Buzz")
    # If not divisible by either, print the number
    else:
        print(num)
```

# 1.1 Common mistakes:

Here are some common mistakes beginners may encounter when solving the FizzBuzz exercise:

1. Incorrect Order of Conditions
Beginners might check divisibility by 3 and 5 separately before checking for both. This leads to incorrect output since the program will print "Fizz" or "Buzz" instead of "FizzBuzz" when the number is divisible by both.

2. Forgetting the else Clause
Some students might forget to include the else statement for printing the number when it is not divisible by 3 or 5

3. Using if instead of elif
Beginners might use multiple if statements instead of elif. This will cause all matching conditions to be evaluated, leading to incorrect results.

4. Incorrect Handling of Range
Some beginners might use the wrong range, such as starting at 0 or ending at 99 instead of 100.

5. Incorrect Use of Modulus Operator
Students may mistakenly use the division operator (/) instead of the modulus operator (%) when checking for divisibility

6. Not Handling Edge Cases
Beginners might forget that 0 is divisible by any number, and it should not be included in the results since the range starts from 1.

7. Hardcoding Numbers Instead of Using Variables
Sometimes students hardcode specific values like 3 and 5 throughout the code. This makes it harder to modify the exercise (e.g., changing the divisors).
Fix: Use variables for the divisors to make the code flexible



# Exercise 1.2 Sum of Digits

Objective: Write a program that takes an integer as input and returns the sum of its digits using a loop.


```python
# Ask the user to input a positive integer
number = int(input("Enter a positive integer: "))

# Initialize the sum of digits to 0
sum_digits = 0

# Use a while loop to calculate the sum of the digits
while number > 0:
    digit = number % 10          # Extract the last digit
    sum_digits += digit          # Add it to the sum
    number = number // 10        # Remove the last digit

# Print the sum of the digits
print("The sum of the digits is:", sum_digits)
```

# 1.2 Common mistakes:

1. Input Type Issues: Forgetting to convert the input from a string to an integer using int(). This can result in a TypeError when trying to perform arithmetic operations on a string.

2. Loop Condition Errors: Using an incorrect loop condition, such as while number >= 0, which leads to an infinite loop when the number reaches 0. The correct condition is while number > 0.

3. Floating-Point Division: Using / instead of // for integer division. / results in a float, causing issues in the loop. Integer division (//) is required to correctly discard the last digit.

4. Negative Numbers: If students input negative numbers, the modulus and division steps may not work as expected. Adding an input validation step for positive integers or using abs(number) to convert negatives to positive might be necessary.

5. Forgetting to Update the Number: Forgetting to reduce the number by dividing it by 10 (number = number // 10) inside the loop, which results in an infinite loop because the number remains the same.

6. Not Handling Edge Cases: Not accounting for special cases like 0 (which should sum to 0) or very large integers, which may expose any weaknesses in logic.

7. Uninitialized Variables: Forgetting to initialize sum_digits to 0, causing either an error or unexpected behavior depending on how Python handles uninitialized variables.

# Exercise 1.3 Compound Interest

Objective: Write a program that calculates the future value of an investment based on a principal amount, an annual interest rate, and the number of years. The calculation should use a loop to apply the interest rate for each year

```python
# Ask the user for the input values
principal = float(input("Enter the principal amount: "))
annual_rate = float(input("Enter the annual interest rate (in %): "))
years = int(input("Enter the number of years: "))

# Convert the annual interest rate from a percentage to a decimal
annual_rate = annual_rate / 100

# Initialize future_value to the principal amount
future_value = principal

# Use a loop to apply the interest rate for each year
for _ in range(years):
    future_value += future_value * annual_rate  # Apply interest for the year

# Print the future value of the investment
print(f"The future value of the investment after {years} years is: {future_value:.2f}")
```

Off-by-One Error in Loop:

4. Wrong loop condition, running the loop one extra or one fewer times. This can lead to incorrect future value results because the compounding happens for too many or too few years.
Floating-Point Precision Issues:

5. Didn't handle the precision of the floating-point numbers correctly. This can lead to outputs showing too many decimal places, making the result look less clean.
Hardcoding Values:

6. Instead of using user input for the principal, interest rate, and number of years, students might hardcode these values. This makes the program less flexible and requires manual changes to test different cases.
Ignoring Edge Cases:

7. overlook edge cases such as:

0 years, where the future value should still be the principal as no compounding occurs.

Without validating the inputs, students might allow negative numbers for years or interest rates, which could lead to unrealistic or incorrect results.

# Exercise 1.4 Analysing Monthly Sales Data

Objective: Write a program that calculates the average monthly sales from a list of sales figures and identifies the highest and lowest sales months.


```python
# Step 1: Create a list of sales figures for each month (hardcoded or input by the user)
sales_figures = [5000, 7000, 8000, 6000, 9000, 7500, 8500, 6700, 7200, 9500, 8700, 7800]

# Step 2: Calculate the total sales using a loop
total_sales = 0
for sales in sales_figures:
    total_sales += sales

# Step 3: Calculate the average sales
average_sales = total_sales / len(sales_figures)

# Step 4: Find the highest and lowest sales figures
highest_sales = max(sales_figures)
lowest_sales = min(sales_figures)

# Step 5: Print the results
print(f"Average monthly sales: {average_sales:.2f}")
print(f"Highest sales month: {highest_sales}")
print(f"Lowest sales month: {lowest_sales}")
```

# Exercise 1.5 Portfolio Performance Analysis

Objective: Write a program that calculates the cumulative returns of a portfolio over a given period, considering both the capital appreciation and the reinvestment of dividends.

```python
# Step 1: Get the user input or hardcode portfolio data
# List of daily/periodic returns as percentages and dividends as percentages of initial capital
capital_appreciation = [0.02, -0.01, 0.015, 0.005, 0.03]  # example values in decimal (2%, -1%, etc.)
dividends = [0.01, 0.005, 0.0, 0.002, 0.008]  # dividends as percentages of initial capital

# Step 2: Initialize starting capital and cumulative return
initial_capital = 10000  # starting capital
current_value = initial_capital

# Step 3: Loop over both the appreciation and dividend data to compute cumulative returns
for i in range(len(capital_appreciation)):
    # Add capital appreciation to the current value
    current_value += current_value * capital_appreciation[i]
    
    # Reinvest dividends (if any)
    current_value += initial_capital * dividends[i]  # dividends reinvested as percentage of initial capital

# Step 4: Calculate total cumulative return
cumulative_return = (current_value - initial_capital) / initial_capital * 100

# Step 5: Print the results
print(f"Initial Capital: ${initial_capital}")
print(f"Final Portfolio Value: ${current_value:.2f}")
print(f"Cumulative Return: {cumulative_return:.2f}%")
```

# Exercise 1.6 Cash Flow Analysis and Net Present Value (NPV) Calculation

Objective: Write a program that calculates the Net Present Value (NPV) of a series of cash flows given a discount rate. The program should also identify which year had the highest and lowest cash inflow.

```python
# Step 1: Define the cash flows for each year (hardcoded or allow user input)
cash_flows = [-10000, 2000, 3000, 4000, 5000, 6000]  # Example: Initial investment is negative (outflow), others are inflows

# Step 2: Define the discount rate (in decimal form)
discount_rate = 0.1  # Example: 10%

# Step 3: Initialize NPV to 0
npv = 0

# Step 4: Calculate NPV using the formula: NPV = Î£ (Cash Flow / (1 + r)^t)
for t in range(len(cash_flows)):
    npv += cash_flows[t] / (1 + discount_rate) ** t

# Step 5: Identify the highest and lowest cash inflows
highest_inflow = max(cash_flows)
lowest_inflow = min(cash_flows)

# Step 6: Identify the year for the highest and lowest inflows (adjust for 0-based indexing)
highest_inflow_year = cash_flows.index(highest_inflow) + 1  # Year index + 1 for human-readable year
lowest_inflow_year = cash_flows.index(lowest_inflow) + 1

# Step 7: Print the results
print(f"Net Present Value (NPV): ${npv:.2f}")
print(f"Highest Cash Inflow: ${highest_inflow} in Year {highest_inflow_year}")
print(f"Lowest Cash Inflow: ${lowest_inflow} in Year {lowest_inflow_year}")
```
