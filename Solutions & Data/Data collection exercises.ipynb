```python
# imports
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import pandas_datareader.data as web
import matplotlib.pyplot as plt
```

## Step 1

```python
# Read the CSV file into a DataFrame
fundamentals_df = pd.read_csv('test_data.csv')
fundamentals_df
```

```python
# Rename the column 'tic' to 'ticker'
fundamentals_df.rename(columns={'tic': 'Ticker'}, inplace=True)
fundamentals_df
```

```python
# Convert datadate to datetime
fundamentals_df['datadate'] = pd.to_datetime(fundamentals_df['datadate']) 
```

```python
# Rename the column 'tic' to 'ticker' (so the same as in share data)
fundamentals_df.rename(columns={'datadate': 'Date'}, inplace=True)
fundamentals_df
```

## Step 2

```python
import yfinance as yf

# Define the ticker symbols
tickers = ['AAPL', 'IBM', 'META']

# Download historical data for the specified tickers from 2010 onwards
data = yf.download(tickers, start='2010-01-01')

# Reset the index to convert the index (date) into a column
data.reset_index(inplace=True)

# Convert the 'Date' column to date format (removing time information)
data['Date'] = data['Date'].dt.date

# Display the first few rows of the resulting DataFrame
data.head()
```

```python
# Flatten the MultiIndex columns
data.columns = [' '.join(col).strip() if isinstance(col, tuple) else col for col in data.columns.values]

# Create a separate DataFrame to hold the long format
long_format = pd.DataFrame()

# Loop through each ticker and extract the relevant data
for ticker in tickers:
    ticker_data = data[['Date', f'Adj Close {ticker}', f'Close {ticker}', f'High {ticker}', f'Low {ticker}', f'Volume {ticker}']]
    ticker_data.columns = ['Date', 'Adj Close', 'Close', 'High', 'Low', 'Volume']  # Rename columns to be the same
    ticker_data['Ticker'] = ticker  # Add a new column for the ticker
    long_format = pd.concat([long_format, ticker_data], ignore_index=True)

# Display the first few rows of the resulting DataFrame
long_format.head()
```

```python
long_format['Date'] = pd.to_datetime(long_format['Date']) 
```

```python
long_format
```

## Step 3

```python
# Sort the data in both dataframe before merging
fundamentals_df.sort_values('Date', inplace=True)
long_format.sort_values('Date', inplace=True)
```

```python
# Use direction='forward' to get the next available daily data on or after the quarterly date
merged_df = pd.merge_asof(fundamentals_df, long_format, on='Date', by='Ticker', direction='forward')
```

```python
merged_df
```

## Step 4

```python
# Calculate ROA, handling division by zero (just calculating one as an example)
merged_df['ROA'] = np.where(merged_df['actq'] != 0, merged_df['niq'] / merged_df['actq'], 0)
```

```python
merged_df
```

## Extra Challenge

### Correlations

```python
# Calculate the correlation between 'Adj Close' and 'ROA'
correlation = merged_df['Adj Close'].corr(merged_df['ROA'])
```

```python
correlation
```

### Visualisations

```python
# Create the scatter plot
plt.figure(figsize=(14, 7))

# Loop through each unique ticker and plot the data
for ticker in merged_df['Ticker'].unique():
    ticker_data = merged_df[merged_df['Ticker'] == ticker]
    plt.scatter(ticker_data['ROA'], ticker_data['Adj Close'], label=ticker, alpha=0.7)

# Adding titles and labels
plt.title('Scatter Plot of ROA vs Adjusted Close by Ticker', fontsize=16)
plt.xlabel('Return on Assets (ROA)', fontsize=12)
plt.ylabel('Adjusted Close Price', fontsize=12)
plt.legend(title='Ticker')
plt.grid()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate the rolling 4-quarter ROA for each ticker
merged_df['Rolling ROA'] = merged_df.groupby('Ticker')['ROA'].rolling(window=4).mean().reset_index(level=0, drop=True)
```

```python
merged_df
```

```python
# Create the plot
plt.figure(figsize=(14, 7))

# Plot Rolling ROA by Ticker
for ticker in merged_df['Ticker'].unique():
    ticker_data = merged_df[merged_df['Ticker'] == ticker]
    plt.plot(ticker_data['Date'], ticker_data['Rolling ROA'], label=f'Rolling ROA - {ticker}')

# Adding titles and labels
plt.title('Rolling 4-Quarter ROA by Ticker', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Rolling ROA', fontsize=12)
plt.legend()
plt.grid()

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
