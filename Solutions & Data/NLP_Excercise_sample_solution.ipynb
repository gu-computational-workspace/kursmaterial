##Step 2: Preprocess the data


```python
import spacy

# Load the medium English model from spaCy
nlp = spacy.load("en_core_web_md")

# Convert text to lowercase
lowercased_text = text.lower()

# Process the text with spaCy's nlp pipeline
doc = nlp(lowercased_text)

# Remove punctuation
no_punctuation_text = ' '.join([token.text for token in doc if not token.is_punct])

# Remove stop words
doc = nlp(no_punctuation_text)
no_stop_words_text = ' '.join([token.text for token in doc if not token.is_stop])

# Lemmatization of the remaining text
doc = nlp(no_stop_words_text)
lemmatized_text = ' '.join([token.lemma_ for token in doc])

# Display the final preprocessed text
lemmatized_text
```

```python
words = lemmatized_text.split()
```

```python
# Custom sentiment dictionary (lemmatized)
sentiment_dict = {
    "positive": [
        "collaborate", "support", "friendly", "teamwork", "opportunity", 
        "develop", "learn", "flexible", "remote", "work-life balance",
        "appreciate", "recognize", "reward", "praise", "enjoy", "love", "engage",
    ],
    "negative": [
        "toxic", "stressful", "unhealthy", "disconnect", "politics", "stagnate", 
        "low pay", "underpaid", "overwhelm", "burnout", "poor leadership", 
        "micromanage", "frustrate", "hate", "boring", "instability", "layoff", 'fight','battle',
    ]
}
```

```python
# Count positive and negative words
positive_count = sum(1 for word in words if word in sentiment_dict['positive'])
negative_count = sum(1 for word in words if word in sentiment_dict['negative'])
```
