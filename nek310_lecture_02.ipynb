<h3>Import the datasets "Gig" and "Spend"</h3>

```python
GIG_data = pd.read_excel('Gig.xlsx', sheet_name='Blad1')

SPEND_data = pd.read_excel('Spend.xlsx', sheet_name='Blad1')
```

<h3>Joining Gig and Spend on Employee_ID and ID</h3>

```python
GIG_SPEND_data=pd.merge(GIG_data, SPEND_data, left_on='EmployeeID', right_on='ID')
```

<h3>Check that join worked as expected</h3>

```python
GIG_SPEND_data.info()
```

<h3>As EmployeeID and ID contain the same data we decide to drop ID</h3>

```python
GIG_SPEND_data.drop(columns=['ID'])
```

<h3>Another option would be to rename EmployeeID as ID or ID as EmployeeID before the join</h3>

```python
GIG_data_2=GIG_data.rename(columns={'EmployeeID':'ID'})
```

```python
GIG_data_2.columns
```

```python
GIG_SPEND_2_data=pd.merge(GIG_data_2, SPEND_data, how="left", on=["ID"])
```

```python
GIG_SPEND_2_data.info()
```

<h3>There are different types of joins, inner and outer, from the left and from the right. Let's see how they differ. In order to display the differences we drop rows with missing values from the datastets.</h3>

```python
GIG_data_3=GIG_data_2.dropna()

SPEND_data_3=SPEND_data.dropna()
```

```python
GIG_data_3.info()
```

```python
SPEND_data_3.info()
```

<h3>Let's see what happens if we do an outer join of GIG and SPEND</h3>

```python
GIG_SPEND_4=GIG_data_3.merge(SPEND_data_3, how='outer', on='ID')

GIG_SPEND_4.info()
```

```python
GIG_SPEND_4.isna().sum()
```

<h3>What about an inner join?</h3>

```python
GIG_SPEND_5=GIG_data_3.merge(SPEND_data_3, how='inner', on='ID')

GIG_SPEND_5.info()
```

```python
GIG_SPEND_5.isna().sum()
```

<h3>So the outer join is like a union and the inner is like an instersection (remember Jaggia & Kelly chapter 4...)</h3>

<h3>What about left and right?</h3>

```python
GIG_SPEND_6=GIG_data_3.merge(SPEND_data_3, how='left', on='ID')

GIG_SPEND_6.info()
```

```python
GIG_SPEND_7=GIG_data_3.merge(SPEND_data_3, how='right', on='ID')

GIG_SPEND_7.info()
```

<h3>Create an indicator variable for wage above the median</h3>

```python
GIG_SPEND_data['Above_median_wage']=(GIG_SPEND_data['Wage']>GIG_SPEND_data['Wage'].median())
```

<h3>Create a crosstable for OwnHome and Above median wage</h3>

```python
pd.crosstab(GIG_SPEND_data['Above_median_wage'],GIG_SPEND_data['OwnHome'])
```

<h3>Descriptive statistics for FoodSpend and TravelSpend grouped by above median wage</h3>

```python
GIG_SPEND_data.groupby(['Above_median_wage'])['TravelSpend'].describe()
```
