```python
def trana_modell(model, xdata, ydata, lr=0.01, epoker=300, skriv_ut_varje=100):
    """
    Trna en PyTorch-modell med SGD-optimering.
    """

    # Optimerare
    optimerare = torch.optim.SGD(model.parameters(), lr=lr)

    # Trningsloop
    for omgng in range(epoker):
        totalt_fel = 0.0
        for x, ytarget in zip(xdata, ydata): 
            xtensor = torch.Tensor([x])
            y = model(xtensor)

            # Berkna felet
            fel = abs(y - ytarget)

            # Summera felet fr kontroll
            totalt_fel += float(fel)

            # Nollstll gradienter fr optimerare
            optimerare.zero_grad()    

            # Berkna gradienter frn felet
            fel.backward()

            # Uppdatera parametrarna i modellen
            optimerare.step()

        # Skriv ut fel ibland fr att flja trningen
        if omgng % skriv_ut_varje == 0:    
            print(f"Omgng {omgng}, fel: {totalt_fel:.4f}")
```
