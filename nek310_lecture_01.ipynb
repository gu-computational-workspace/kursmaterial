<h3>Make a histogram for Jobs</h3>

```python
import seaborn as sns 
sns.set_theme() 

import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10,7) 

GIG_data['Job'].value_counts().plot(kind='bar', x='Job')
plt.title("Distribution of Jobs",fontsize=20) 
plt.tick_params(axis='both', which='major', labelsize=20) 
plt.xlabel('Job',fontsize=16) 
plt.ylabel('Number of workers',fontsize=16) 
#plt.xticks(fontsize=16,ha="left") 
plt.xticks(fontsize=16,rotation=-45,ha="left")
plt.yticks(fontsize=16) 
plt.show()
```

<h3>Jobs grouped on Industries</h3>

```python
pd.crosstab(GIG_data['Industry'],GIG_data['Job']).plot(kind='bar',stacked=False)


plt.title("Distribution of jobs",fontsize=20) 
plt.tick_params(axis='both', which='major', labelsize=20) 
plt.xlabel('Industry',fontsize=24) 
plt.ylabel('Numbers of workers',fontsize=16) 
#plt.xticks(fontsize=16,ha="left") 
plt.xticks(fontsize=16,rotation=0, ha="center")
plt.yticks(fontsize=16) 
plt.legend(fontsize=16) 
plt.show()
```
